### Instructor Demo: Indexing

This program reads performs several slice and dice operations through indexing via loc and iloc functions.

```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
```

### Read CSV in as DataFrame

```python
# Set the file path
file_path = Path('../Resources/people.csv')

# Read in the CSV as a DataFrame
people_csv = pd.read_csv(file_path)
people_csv.head()
```

```python
people_csv.describe()
```

```python
people_csv.describe(include="all")
```

```python
#take the first record and give it back to us as a series(similar to a dict)
people_csv.iloc[0]
```

```python
people_csv.iloc[:10]
```

```python
people_csv.iloc[-1]
```

```python
#go from 0 to 3 including only the zero column
people_csv.iloc[0:3,0].head()
```

```python
#go from 0 to 3 including only the [2] column
people_csv.iloc[0:3,2].head()
```

```python
# go from 1 to 2 including the first 4 columns
people_csv.iloc[1:2, range(4)]
```

```python
#assign this to a dataframe
dataframe = people_csv.iloc[[0, 2, 3], [3, 5]]
```

```python
dataframe.head()
```
